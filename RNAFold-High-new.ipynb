```python
import pandas as pd
import RNA
```

```python
data = pd.read_csv("with_high_confidence_score_gc_sequences_dec_2025_with_energy.csv")
```

```python
for idx,row in data.iterrows():
    
    seq1 = row['Forward_strand']
    fc1 = RNA.fold_compound(seq1)
    (ss,mfe) = fc1.mfe()
    data.loc[idx,'Forward_strand_sec_struc'] = ss
    data.loc[idx,'ΔGforward'] = mfe



    seq2 = row['Reverse_strand']
    fc2 = RNA.fold_compound(seq2)
    (ss2,mfe2) = fc2.mfe()
    data.loc[idx,'Reverse_strand_sec_struc'] = ss2
    data.loc[idx,'Δgreverse '] = mfe2

```
