```python
import pandas as pd
import subprocess
```

```python
data = pd.read_csv("/data/shuba/Lab_work/genomics/with_high_confidence_score_gc_sequences_dec_2025_with_energy.csv")
```

```python
for idx,row in data.iterrows():
    
    f_seq = row['Forward_strand'].replace('T','U')
    r_seq = row['Reverse_strand'].replace('T','U')

    #DUPLEX
    #anterior
    with open("file1",'w') as file1:
        file1.write(f_seq)
    with open("file2",'w') as file2:
        file2.write(r_seq)


    subprocess.run(["./DuplexFold","file1", "file2", "output"])

    with open("output",'r') as f:
        for line in f.readlines():
            print(line)
            # print("anterior: ",line.split()[1]+line.split()[2]+line.split()[3])
            data.loc[idx,'Δgduplex']=line.replace('\n','')
            # .split()[1b]+line.split()[2]+line.split()[3]
            break

    
```

```python
for idx,row in data.iterrows():
    
    f_seq = row['Forward_strand'].replace('T','U')
    r_seq = row['Reverse_strand'].replace('T','U')
#anterior
    with open("file1",'w') as file1:
        file1.write(f_seq[0:7])
    with open("file2",'w') as file2:
        file2.write(r_seq[0:7])


    subprocess.run(["./DuplexFold","file1", "file2", "output"])

    with open("output",'r') as f:
        # print(f.readlines())
        for line in f.readlines():
            print(line)
            # print("anterior: ",line.split()[1]+line.split()[2]+line.split()[3])
            data.loc[idx,'Δganterior']=line.replace('\n','')
            # .split()[1]+line.split()[2]+line.split()[3]
            break
```

```python
for idx,row in data.iterrows():
    
    f_seq = row['Forward_strand'].replace('T','U')
    r_seq = row['Reverse_strand'].replace('T','U')
    #middle
    with open("file1",'w') as file1:
        file1.write(f_seq[7:17])
    with open("file2",'w') as file2:
        file2.write(r_seq[7:17])

    subprocess.run(["./DuplexFold","file1", "file2", "output"])

    with open("output",'r') as f:
        for line in f.readlines():
            print(line)
            # print("middle: ",line.split()[1]+line.split()[2]+line.split()[3])
            data.loc[idx, 'Δgmiddle'] = line.replace('\n','')
            # split()[1]+line.split()[2]+line.split()[3]
            break
    

```

```python
for idx,row in data.iterrows():
    
    f_seq = row['Forward_strand'].replace('T','U')
    r_seq = row['Reverse_strand'].replace('T','U')
    #posterior
    with open("file1",'w') as file1:
        file1.write(f_seq[17:24])
    with open("file2",'w') as file2:
        file2.write(r_seq[17:24])

    subprocess.run(["./DuplexFold","file1", "file2", "output"])

    with open("output",'r') as f:
        # print(f.readlines())
        for line in f.readlines():
            print(line)
            # print("posterior: ",line.split()[1]+line.split()[2]+line.split()[3])
            data.loc[idx, 'Δgposterior'] = line.replace('\n','')
            # .split()[1]+line.split()[2]+line.split()[3]
            break
```
